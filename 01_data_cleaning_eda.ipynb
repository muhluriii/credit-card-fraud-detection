```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For better visual formatting
sns.set(style='whitegrid')
%matplotlib inline

# Step 2: Load Dataset
df = pd.read_csv('creditcard.csv')
df.shape, df.head()

# Data overview
df.info()
df.describe()
df.isnull().sum()
df['Class'].value_counts()
df['Class'].value_counts(normalize=True)

plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title('Fraud vs Legit Transactions')
plt.xticks([0, 1], ['Legit (0)', 'Fraud (1)'])
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(12,5))

# Distribution of legit transaction amounts
sns.histplot(df[df['Class'] == 0]['Amount'], bins=100, color='blue', label='Legit')
sns.histplot(df[df['Class'] == 1]['Amount'], bins=100, color='red', label='Fraud')
plt.legend()
plt.title('Transaction Amounts: Legit vs Fraud')
plt.xlabel('Amount ($)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(20, 15))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()

```
